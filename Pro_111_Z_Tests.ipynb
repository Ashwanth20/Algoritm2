```python
import plotly.figure_factory as ff
import plotly.graph_objects as go
import pandas as pd
import statistics
import random
import csv

df = pd.read_csv("medium_data.csv")
data = df["reading_time"].tolist()
```

```python
fig = ff.create_distplot([data], ["Math Scores"], show_hist = False)
fig.show()
```

```python
mean = statistics.mean(data)
std_deviation = statistics.stdev(data)
print("Mean of Population : ", mean)
print("Standard Deviation of Population : ", std_deviation)
```

```python
def random_set_of_mean(counter):
    dataset = []
    for i in range(0,counter):
        random_index = random.randint(0,len(data) - 1)
        value = data[random_index]
        dataset.append(value)
    mean = statistics.mean(dataset)
    return mean
```

```python
mean_list = []
for i in range(0,1000):
    set_of_means = random_set_of_mean(100)
    mean_list.append(set_of_means)
```

```python
std_deviation = statistics.stdev(mean_list)
mean = statistics.mean(mean_list)
print("Mean of sampling Distribution: ", mean)
print("Standard Deviation of sampling Distribution: ", std_deviation)
```

```python
fig = ff.create_distplot([mean_list], ["student marks"], show_hist = False)
fig.add_trace(go.Scatter(x = [mean,mean], y = [0,0.20], mode = 'lines', name = 'Mean'))
fig.show()
```

```python
first_std_dev_start, first_std_dev_end = mean - std_deviation, mean + std_deviation
sec_std_dev_start, sec_std_dev_end = mean - (2*std_deviation), mean + (2*std_deviation)
third_std_dev_start, third_std_dev_end = mean - (3*std_deviation), mean + (3*std_deviation)
print("std_1", first_std_dev_start, first_std_dev_end)
print("std_2", sec_std_dev_start, sec_std_dev_end)
print("std_3", third_std_dev_start, third_std_dev_end)
```

```python
fig = ff.create_distplot([mean_list], ["student marks"], show_hist = False)
fig.add_trace(go.Scatter(x = [mean,mean], y = [0,0.17], mode = 'lines', name = 'Mean'))
fig.add_trace(go.Scatter(x = [first_std_dev_start,first_std_dev_start], y = [0,0.17], mode = 'lines', name = 'Standard Deviation I Start'))
fig.add_trace(go.Scatter(x = [first_std_dev_end,first_std_dev_end], y = [0,0.17], mode = 'lines', name = 'Standard Deviation I End'))
fig.add_trace(go.Scatter(x = [sec_std_dev_start,sec_std_dev_start], y = [0,0.17], mode = 'lines', name = 'Standard Deviation II Start'))
fig.add_trace(go.Scatter(x = [sec_std_dev_end,sec_std_dev_end], y = [0,0.17], mode = 'lines', name = 'Standard Deviation II End'))
fig.add_trace(go.Scatter(x = [third_std_dev_start,third_std_dev_start], y = [0,0.17], mode = 'lines', name = 'Standard Deviation III Start'))
fig.add_trace(go.Scatter(x = [third_std_dev_end,third_std_dev_end], y = [0,0.17], mode = 'lines', name = 'Standard Deviation III End'))
fig.show()
```
